```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
g = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2 = g.text
count = 0

data = json.loads(response)
data2 = json.loads(response2)
print('這個 API 一次會回傳幾筆資料:',len(data)) 
for d in data:
    count = count + 1
print('這個 API 一次會回傳幾筆資料:',count)

get_type = type(data)
print(get_type)

count1 = 0
count2 = 0


for d in data:
    while count1 <1:
        count1 = count1 + 1
        for x in d.keys():
            count2 = count2 + 1
            print('欄位:',x)
print('總欄位:',count2)            
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
data = json.loads(response)
data2 = json.loads(response2)
count = 0
count1 = 0
messageCount = 0
goodCount = 0
messageCount1 = 0
goodCount1 = 0

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
    messageCount = d['commentCount'] + messageCount
    goodCount = d['likeCount'] + goodCount
    count = count + 1

for g in data2:
    print(g['title'])
    print(g['createdAt'])
    print(g['commentCount'])
    print(g['likeCount'])
    messageCount1 = g['commentCount'] + messageCount1
    goodCount1 = g['likeCount'] + goodCount1
    count1 = count1 + 1
```

```python
print('熱門文章「平均數」:',count)
print('非熱門文章「平均數」:',count1)
print('熱門文章「留言數」:',messageCount/count)
print('熱門文章「按讚數」:',goodCount/count)
print('非熱門文章「留言數」:',messageCount1)
print('非熱門文章「按讚數」:',goodCount1)
print('熱門文章「平均留言人數」:',messageCount/count)
print('熱門文章「平均按讚人數」:',goodCount/count)
print('非熱門文章「平均留言人數」:',messageCount1/count1)
print('非熱門文章「平均按讚人數」:',goodCount1/count1)
```
