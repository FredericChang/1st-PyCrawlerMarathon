### ① 取出今天所有的發文

```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
dict = {}
articles = []

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    articles.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    print(d.find(class_="tag").text)
    key = d.find(class_="tag").text
    if key in dict.keys(): 
        valueAdd = dict.get(key)
        value = valueAdd + 1
        #dict[key] = 1
        dict[key] = value

        #print("Present, ", end =" ") 
        #print("value =", dict[key]) 
    else: 
        dict[key] = 1 
```

### ③ 哪一個類別的文章最多

```python
import operator
max(dict.items(), key=operator.itemgetter(1))[0]
```

```python
columns = ['日期', '類別', '標題']
df = pd.DataFrame(articles, columns=columns)
df
```

```python

df2 = pd.DataFrame(df.groupby('類別').size(), columns=['文章數量'])
df2['排名'] = df2.rank(method='min', ascending=False)
df2.sort_values(by=['排名'])
```
